7) In the case of simple linear regression of Y onto X, the R^2 statistic equals the square of correlation between X and Y.

The R^2 statistic is given by

R^2 = 1 - RSS/TSS

The correlation between X and Y is given by

Cor(X,Y) = $\frac{\sum_{i=1}^{n}(x_i- x)(y_i - y)}{\sqrt{\sum_{i=1}^{n}(x_i- x)^2} * \sqrt{\sum_{i=1}^{n}(y_i- y)^2}}$

where x, y are mean values. If we now have, x, y as 0, we will get Cor(X,Y) as 

Cor(X,Y) = $\frac{\sum_{i=1}^{n}(x_i)(y_i)}{\sqrt{\sum_{i=1}^{n}(x_i)^2} * \sqrt{\sum_{i=1}^{n}(y_i)^2}}$

Now, R^2 is given by

 R^2 = 1 - RSS/TSS
 
 = 1 - $\frac{\sum_{i=1}^{n}(y_i - yhat)^2}{\sum_{i=1}^{n}(y_i - ybar)^2}$
 = 1 - $\frac{\sum_{i=1}^{n}(y_i - (\beta0 + \beta1*x_i))^2}{\sum_{i=1}^{n}(y_i - ybar)^2}$
 
 Now $\beta$0 = 0, $\beta$1 = $\frac{\sum_{i=1}^{n}(x_i)(y_i)}{\sum_{i=1}^{n}(x_i)^2}$
 
 so, R^2 is
 
 = $\frac{\sum_{i=1}^{n}(y_i)^2 - \sum_{i=1}^{n}(y_i)^2 + 2\sum_{i=1}^{n}(\beta1*x_i)(y_i) - \sum_{i=1}^{n}(\beta1*x_i)^2}{\sum_{i=1}^{n}(y_i)^2}$
  = $\frac{ 2\sum_{i=1}^{n}(\beta1*x_i)(y_i) - \beta1^2*\sum_{i=1}^{n}(x_i)^2}{\sum_{i=1}^{n}(y_i)^2}$
  = $\frac{\sum_{i=1}^{n}(x_i)(y_i)}{\sum_{i=1}^{n}(y_i)^2\sum_{i=1}^{n}(x_i)^2}$
    
    
which is square 0f Cor(X,Y)
