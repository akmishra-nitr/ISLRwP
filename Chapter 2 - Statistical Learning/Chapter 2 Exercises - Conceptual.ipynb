    5 Advantages of a very flexible approach for regression or classification over a less flexible approach - low bias, model can explain training data to a large extent, better for prediction. Disdvantages - chances of overfitting, needs a lot of observations, not so good for inference as it can be very complex to explain the relationship of between predictors and response. 
    
    Less flexible approach favorable when inference is the goal and number of observations is not large enough.

    6 A parametric approach simplifies the problem from identifying the function to predicting the parameters based on the observations. We can then simply focus on predicting the response with high accuracy. 

    Non-parametric approach, on the other hand, tries to find out the form ofthe function. Hence, it is much more useful in inference problems.
